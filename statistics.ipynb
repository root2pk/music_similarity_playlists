## Jupyter notebook to plot statistical distributions of the music collection

```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# Read discogs metadata json file to get the genre corresponding to each index
with open('metadata/discogs-effnet-bs64-1.json') as file:
    metadata_dict = json.load(file)

genre_list = metadata_dict["classes"]
```

```python
# Reduce genre list based on parent genre, i.e. the genre before the -- in each value
reduced_genre_list = []
for genre in genre_list:
    reduced_genre_list.append(genre.split('--')[0])
```

```python
# Read genre_predictions.csv and plot distribution of genres
df = pd.read_csv('genre_predictions.csv', usecols = [2])
df.columns = ['genre_numbers']
list_of_genres = df['genre_numbers'].tolist()
# Convert genre numbers to genre names
genre_names = [genre_list[i] for i in list_of_genres]
print(genre_names)
```

```python
# Plot tempo distribution
df = pd.read_csv('features.csv', usecols = [1])
df.columns = ['Tempo']
df['Tempo'].plot(kind='hist', bins=20)
plt.title('Tempo distribution')
plt.xlabel('Tempo')
plt.ylabel('Number of songs')
plt.show()
```

```python
# Plot danceability distribution
df = pd.read_csv('features.csv', usecols = [11])
df.columns = ['Danceability']
df['Danceability'].plot(kind='hist', bins=20)
plt.title('Danceability distribution')
plt.xlabel('Danceability')
plt.ylabel('Number of songs')
plt.show()
```
