## Jupyter notebook to plot statistical distributions of the music collection

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read genre_predictions.csv and plot distribution of genres usng seaborn
df = pd.read_csv('genre_predictions.csv', usecols = [4])
df.columns = ['Parent Genre']
sns.displot(data=df, x='Parent Genre')
sns.rugplot(data=df, x='Parent Genre')
plt.title('Distribution of Parent Genres')
plt.xticks(rotation=45)
plt.xlabel(None)
plt.show()
```

```python
# Extract specific genre predictions from genre_predictions.csv
df = pd.read_csv('genre_predictions.csv', usecols = [3])
df.columns = ['Genre']
# Save the extracted genre predictions and number of occurrences to a new tsv file
df['Genre'].value_counts().to_csv('genre_counts.tsv', sep='\t', header = ['Count'])
```

#### Tempo distribution

```python
# Plot tempo distribution
df = pd.read_csv('features.csv', usecols = [1])
df.columns = ['Tempo']
sns.histplot(data=df, x='Tempo', kde=True)
sns.rugplot(data=df, x='Tempo')
plt.title('Tempo distribution')
plt.xlabel('Tempo (bpm)')
plt.ylabel('Number of tracks')
plt.show()
```

##### Arousal and Valence

```python
# Plot arousal and valence in a 2D distribution
df = pd.read_csv('features.csv', usecols=[11, 12])
df.columns = ['Arousal', 'Valence']

# Subtract 5 from both 'Arousal' and 'Valence' columns, since range is 1 - 9
df['Arousal'] = df['Arousal'] - 5
df['Valence'] = df['Valence'] - 5

# Plot the scatter plot
ax = sns.scatterplot(data=df, x='Valence', y='Arousal')

plt.title('Arousal and Valence distribution', pad=15)
plt.xlabel('Valence')
plt.ylabel('Arousal')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axvline(x=0, color='r', linestyle='--')
plt.axhline(y=0, color='r', linestyle='--')

# Add labels
ax.text(4, 0, 'positive', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))
ax.text(-4, 0, 'negative', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))
ax.text(0, 4, 'high', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))
ax.text(0, -4, 'low', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))
ax.text(0, 0, 'neutral', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))

plt.show()
```
