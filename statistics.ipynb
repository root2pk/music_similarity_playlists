## Jupyter notebook to plot statistical distributions of the music collection

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read genre_predictions.csv and plot distribution of genres
df = pd.read_csv('genre_predictions.csv', usecols = [4])
df.columns = ['Parent Genre']
df['Parent Genre'].value_counts().plot(kind='bar')
plt.title('Distribution of Parent Genres')
plt.xticks(rotation=45)
plt.xlabel('Parent Genre')
plt.ylabel('Count')
plt.show()
```

#### Tempo distribution

```python
# Plot tempo distribution
df = pd.read_csv('features.csv', usecols = [1])
df.columns = ['Tempo']
df['Tempo'].plot(kind='hist', bins=20)
plt.title('Tempo distribution')
plt.xlabel('Tempo (bpm)')
plt.ylabel('Number of songs')
plt.show()
```

#### Danceability distribution

```python
# Plot danceability distribution
df = pd.read_csv('features.csv', usecols = [10])
df.columns = ['Danceability']
df['Danceability'].plot(kind='hist', bins=20)
plt.title('Danceability distribution')
plt.xlabel('Danceability (1-10)')
plt.ylabel('Number of songs')
plt.show()
```

#### Key and Scale distribution
##### 1) Temperley

```python
# Plot key and scale distribution - Temperley
df = pd.read_csv('features.csv', usecols = [2,3])
df.columns = ['key', 'scale']
# Combine key and scale into one column
df['Key and Scale'] = df['key'] + ' ' + df['scale']
df['Key and Scale'].value_counts().plot(kind='bar')
plt.title('Key and Scale distribution - Temperley')
plt.xlabel('Key and Scale')
plt.ylabel('Number of songs')
plt.show()
```

#### Loudness Distribution

```python
# Plot loudness distribution
df = pd.read_csv('features.csv', usecols = [8])
df.columns = ['Loudness']
df['Loudness'].plot(kind='hist', bins=20)
plt.title('Loudness distribution')
plt.xlabel('Loudness (LUFS)')
plt.ylabel('Number of songs')
plt.show()
```

##### Arousal and Valence

```python
# Plot arousal and valence in a 2D distribution
df = pd.read_csv('features.csv', usecols=[11, 12])
df.columns = ['Arousal', 'Valence']

# Subtract 5 from both 'Arousal' and 'Valence' columns, since range is 1 - 9
df['Arousal'] = df['Arousal'] - 5
df['Valence'] = df['Valence'] - 5

# Plot the scatter plot
ax = df.plot(kind='scatter', x='Valence', y='Arousal')

plt.title('Arousal and Valence distribution', pad=15)
plt.xlabel('Valence')
plt.ylabel('Arousal')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axvline(x=0, color='r', linestyle='--')
plt.axhline(y=0, color='r', linestyle='--')

# Add labels
ax.text(4, 0, 'positive', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))
ax.text(-4, 0, 'negative', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))
ax.text(0, 4, 'high', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))
ax.text(0, -4, 'low', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))
ax.text(0, 0, 'neutral', ha='center', va='center', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round'))

plt.show()
```
