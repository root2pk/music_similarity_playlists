```python
# Initialse features.csv and genre_predicitons.csv and clear them
open('features.csv', 'w').close()
open('genre_predictions.csv', 'w').close()
```

```python
# Loop through all audio files
for audio_file in tqdm(audio_files[:10]):
    # Load audio file
    audio_stereo, audio_mono  = m.load_audio_file(audio_file) 
    # Extract features
    ess.extract_features(audio_mono, audio_stereo)
    # Genre predictions dictionary
    genre_predictions = ess.write_genre_dict(audio_file)
    # Features dictionary
    features_dict = ess.write_features_dict(audio_file)
    # Save genre predictions to csv
    df = pd.DataFrame([genre_predictions])
    df.to_csv('genre_predictions.csv',mode = 'a', header = False, index = False)
    # Save the rest of the features to csv
    df = pd.DataFrame([features_dict])
    df.to_csv('features.csv',mode = 'a', header = False, index = False)
```
